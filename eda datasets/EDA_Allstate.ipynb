```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

# Load data and basic understanding of data frames

```python
train = pd.read_csv('../data/Allstate/train.csv')
test = pd.read_csv('../data/Allstate/test_v2.csv')
```

```python
train[['customer_ID', 'shopping_pt', 'record_type', 'day', 'time', 'A', 'B', 'C', 'D', 'E', 'F', 'G']].head(60)
```

```python
test[['customer_ID', 'shopping_pt', 'record_type', 'day', 'time', 'A', 'B', 'C', 'D', 'E', 'F', 'G']].head(60)
```

```python
print('Num total interactions train set: ' + str(len(train)))
print('Num total interactions test set: ' + str(len(test)))
print('Num unique customers train set: ' + str(len(train.customer_ID.unique())))
print('Num unique customers test set: ' + str(len(test.customer_ID.unique())))
```

```python
grouped = train.groupby('customer_ID')
days_session = []
for name,group in grouped:
    max_day = group['day'].max()
    min_day = group['day'].min()
    days_session.append(max_day - min_day)
    group[['A', 'B', 'C', 'D', 'E', 'F', 'G']]
    group[[]]
    break
```

```python
print('Mean days interaction in test: ' + np.mean(days_session))
```
